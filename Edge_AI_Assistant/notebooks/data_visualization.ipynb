### Matplotlib & Seaborn - Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample Data
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [10, 20, 15, 25]
    }
df = pd.DataFrame(data)

# Basic Bar Plot
plt.figure(figsize=(6,4))
plt.bar(df['Category'], df['Values'], color='skyblue')
plt.xlabel("Category")
plt.ylabel("Values")
plt.title("Bar Plot Example")
plt.show()

# Seaborn Histogram
sns.histplot(df['Values'], bins=5, kde=True)    # kde=True for Kernel Density Estimation
plt.title("Seaborn Histogram")
plt.show()
```

```python
import numpy as np

# histogram
data = [1, 1, 1, 2, 2, 3, 3, 3, 3, 4, 5, 5, 5, 5, 5]
plt.hist(data, bins = 5, color = 'skyblue', edgecolor = 'black')
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

```python
# scatter plot
x = np.arange(1,21) # 1 to 20
y = np.random.randint(1, 20, 20)    # random integers between 1 to 20
plt.figure(figsize=(6,4))
plt.scatter(x, y, color='skyblue')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")
plt.show()

# line plot
x = np.arange(1, 21)
y = np.random.randint(1, 20, 20)
plt.figure(figsize=(6,4))
plt.plot(x, y, color='red')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot Example")
plt.show()
```

```python
# bar plot
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [10, 20, 15, 25]
    }
df = pd.DataFrame(data)
plt.bar(df['Category'], df['Values'], color='skyblue')
plt.xlabel("Category")
plt.ylabel("Values")
plt.title("Bar Plot Example")
plt.show()
```

```python
# pie chart
sizes = [10, 20, 30, 40]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)   # autopct to show percentage on pie chart slices and startangle to rotate the pie chart by 140 degrees anticlockwise from the x-axis. Also, 1.1f% is used to show one decimal place.
plt.axis('equal')   # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Pie Chart Example")
plt.show()
```

```python
# box plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]   # 3 different normal distributions with mean 0 and standard deviation 1, 2, and 3 respectively.
plt.boxplot(data, vert=True, patch_artist= True)    # vert=True to plot the box plot vertically and patch_artist=True to fill the box plot with color.
plt.xlabel("standard deviation")
plt.ylabel("normal distribution values")
plt.title("Box Plot Example")
plt.show()

# violin plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.violinplot(data, vert=True, showmedians=True, showmeans=True)    # vert=True to plot the violin plot vertically, showmedians=True to show the median line, and showmeans=True to show the mean line.
plt.xlabel("standard deviation")
plt.ylabel("normal distribution values")
plt.title("Violin Plot Example")
plt.show()
```

```python
# heatmap
data = np.random.rand(5,5)   # 10x10 random data
sns.heatmap(data, cmap='coolwarm', annot=True)    # cmap='coolwarm' to set the color map and annot=True to show the values on the heatmap.
plt.title("Heatmap Example")
plt.show()
```

```python
# pairplot
data = sns.load_dataset('iris')    # load iris dataset
sns.pairplot(data, hue='species')   # hue='species' to color the data points based on the species column.
plt.title("Pairplot Example")
plt.show()
```

```python
# Customize plots for better interpretability

# Customize Bar Plot
plt.figure(figsize=(6,4))
plt.bar(df['Category'], df['Values'], color='skyblue')
plt.xlabel("Category")
plt.ylabel("Values")
plt.title("Bar Plot Example")
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees
plt.show()

# Customize Histogram
sns.histplot(df['Values'], bins=5, kde=True)
plt.title("Seaborn Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

# Customize Scatter Plot
x = np.arange(1,21)
y = np.random.randint(1, 20, 20)
plt.figure(figsize=(6,4))
plt.scatter(x, y, color='skyblue')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")
plt.xticks(np.arange(0, 21, 2))  # Set x-axis ticks to show every 2 units
plt.yticks(np.arange(0, 21, 2))  # Set y-axis ticks to show every 2 units
plt.show()

# Customize Line Plot
x = np.arange(1, 21)
y = np.random.randint(1, 20, 20)
plt.figure(figsize=(6,4))
plt.plot(x, y, color='red')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot Example")
plt.xticks(np.arange(0, 21, 2))  # Set x-axis ticks to show every 2 units
plt.yticks(np.arange(0, 21, 2))  # Set y-axis ticks to show every 2 units
plt.show()

# Customize Pie Chart
sizes = [10, 20, 30, 40]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title("Pie Chart Example")
plt.show()

# Customize Box Plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# display box plot with different colors
box = plt.boxplot(data, vert=True, patch_artist= True)
colors = ['pink', 'lightblue', 'lightgreen']
for patch, color in zip(box['boxes'], colors):  # zip() function is used to iterate over two lists simultaneously.
    patch.set_facecolor(color)
plt.xlabel("standard deviation")
plt.ylabel("normal distribution values")
plt.title("Box Plot Example")
plt.show()


```
